# Modelos de ML

```python
import pandas as pd
import matplotlib
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import hamming_loss
from sklearn.metrics import precision_recall_fscore_support
```

```python
pd.set_option('display.max_columns', None)
df_model = pd.read_csv('df_esteroids_MB_IN_AR.csv', sep=";")
#df_model.drop(['Unnamed: 0'], axis= 1, inplace=True)
df_model
```

## Split train and test

```python
yi= df_model.iloc[:,0]
yi
```

```python
# Dummy variabels
str_yi = [str(x) for x in df_model.iloc[:,0]]
yi = pd.DataFrame(str_yi)
yi_dummy = pd.get_dummies(yi)
yi_dummy
```

```python
# fazer comite machine

# separar em 2 grupos 
## 1 grupo = 0 , 2 , 4
## 2 grupo = 0, 5, 1, 3, 6

# 100
# 010
# 001
# 000
```

```python
# separando grupo 1 e 2
df_grup1 =[]
df_grup2 =[]
for f in range(len(df_model.iloc[:,0])):
    if df_model.iloc[f,0]==2 or df_model.iloc[f,0]==0 or df_model.iloc[f,0]==4:
        df_grup1.append(df_model.iloc[f,:]) 
    if df_model.iloc[f,0]==0 or df_model.iloc[f,0]==5 or df_model.iloc[f,0]==1 or df_model.iloc[f,0]==3 or df_model.iloc[f,0]==6:
        df_grup2.append(df_model.iloc[f,:]) 
df_grup1 = pd.DataFrame(df_grup1)
df_grup2 = pd.DataFrame(df_grup2)
```

```python
df_grup2
```

```python
xi = df_grup1.iloc[:,1:27]
yi = df_grup1.iloc[:,0]
yi
```

```python
yi_str = [str(x) for x in yi]
yi_str = pd.DataFrame(yi_str)
yi_dummy = pd.get_dummies(yi_str)
yi_dummy
```

```python
X_train, X_test, y_train, y_test = train_test_split(xi,yi_dummy, random_state=2)
```

```python
y_train
```

## Random Forest Classifier Tuning

```python
RFC = RandomForestClassifier(random_state=1, n_estimators = 70)
```

```python
model_RFC = RFC.fit(X_train,y_train)
```

```python
prediction_RFC = model_RFC.predict(X_test)
```

```python
mc = confusion_matrix(y_test,prediction_RFC)
mc
```

```python
plot_confusion_matrix(RFC, X_test, y_test)  
```

```python
# accuracy
accuracy_RFC = accuracy_score(y_test, prediction_RFC)
accuracy_RFC
```

```python
# MCC
mcc_RFC = matthews_corrcoef(y_test, prediction_RFC)
mcc_RFC

```

```python
# hamming_loss
hamming_loss_RFC = hamming_loss(y_test, prediction_RFC)
hamming_loss_RFC
```

```python
# precision
precision_recall_fscore_support(y_test, prediction_RFC, average='macro')
```

```python
# tuning n_estimators

for i in range(5,1000):
    RFC = RandomForestClassifier(random_state=1, n_estimators = i)
    model_RFC = RFC.fit(X_train,y_train)
    prediction_RFC = model_RFC.predict(X_test)
    accuracy_RFC = accuracy_score(y_test, prediction_RFC)
    #mcc_RFC = matthews_corrcoef(y_test, prediction_RFC)
    hamming_loss_RFC = hamming_loss(y_test, prediction_RFC)
    if hamming_loss_RFC < 0.2:
         print("random_state = {i}, accuracy : {acc}, hamming_loss : {hl}".format(acc = accuracy_RFC, hl = hamming_loss_RFC, i = i))
     
    
```

#### Best n_estimator=108 or 7

```python
# tuning max_depth

for i in range(1,500):
    RFC = RandomForestClassifier(random_state=1, n_estimators = 29, max_depth = i)
    model_RFC = RFC.fit(X_train,y_train)
    prediction_RFC = model_RFC.predict(X_test)
    accuracy_RFC = accuracy_score(y_test, prediction_RFC)
    #mcc_RFC = matthews_corrcoef(y_test, prediction_RFC)
    hamming_loss_RFC = hamming_loss(y_test, prediction_RFC)
    if hamming_loss_RFC < 0.2:
        print("max_depth = {i}, accuracy : {acc}, hamming_loss : {hl}, ".format(acc = accuracy_RFC, hl = hamming_loss_RFC, i = i))
```

#### Best max_depth = 13 or 10

```python
# tuning min_samples_split

for i in range(2,30):
    RFC = RandomForestClassifier(random_state=1, n_estimators = 7, max_depth = 9, min_samples_split = i)
    model_RFC = RFC.fit(X_train,y_train)
    prediction_RFC = model_RFC.predict(X_test)
    accuracy_RFC = accuracy_score(y_test, prediction_RFC)
    #mcc_RFC = matthews_corrcoef(y_test, prediction_RFC)
    hamming_loss_RFC = hamming_loss(y_test, prediction_RFC)
    if hamming_loss_RFC < 0.2:
        print("min_samples_split = {i}, accuracy : {acc}, hamming_loss : {hl}".format(acc = accuracy_RFC, hl = hamming_loss_RFC, i = i))
```

#### Best min_samples_split  = 2

```python
# tuning min_samples_leaf
for i in range(1,40):
    RFC = RandomForestClassifier(random_state=1, n_estimators = 7, max_depth = 10, min_samples_split = 2, min_samples_leaf = i)
    model_RFC = RFC.fit(X_train,y_train)
    prediction_RFC = model_RFC.predict(X_test)
    accuracy_RFC = accuracy_score(y_test, prediction_RFC)
    #mcc_RFC = matthews_corrcoef(y_test, prediction_RFC)
    hamming_loss_RFC = hamming_loss(y_test, prediction_RFC)
    if accuracy_RFC > 0.63:
        print("min_samples_leaf = {i}, accuracy : {acc}, hamming_loss : {hl}".format(acc = accuracy_RFC, hl = hamming_loss_RFC, i = i))
```

#### Best min_samples_leaf = 1

```python
# tuning random_state
for i in range(1,1500):
    RFC = RandomForestClassifier(random_state=i, n_estimators = 7, max_depth = 10, min_samples_split = 2, min_samples_leaf = 10)
    model_RFC = RFC.fit(X_train,y_train)
    prediction_RFC = model_RFC.predict(X_test)
    accuracy_RFC = accuracy_score(y_test, prediction_RFC)
    mcc_RFC = matthews_corrcoef(y_test, prediction_RFC)
    hamming_loss_RFC = hamming_loss(y_test, prediction_RFC)
    if mcc_RFC > 0.65:
        print("random_state = {i}, accuracy : {acc}, mcc: {mcc}, hamming_loss : {hl}".format(mcc=mcc_RFC, acc = accuracy_RFC, hl = hamming_loss_RFC, i = i))

```

```python
# Search seed
for i in range(1,15000):
    X_train, X_test, y_train, y_test = train_test_split(xi,yi, random_state=i)
    RFC = RandomForestClassifier(random_state=742, n_estimators = 108, max_depth = 13, min_samples_split = 2, min_samples_leaf = 1)
    model_RFC = RFC.fit(X_train,y_train)
    prediction_RFC = model_RFC.predict(X_test)
    accuracy_RFC = accuracy_score(y_test, prediction_RFC)
    hamming_loss_RFC = hamming_loss(y_test, prediction_RFC)
    mcc_RFC = matthews_corrcoef(y_test, prediction_RFC)
    if mcc_RFC > 0.77:
        print("random_state = {i}, accuracy : {acc},mcc: {mcc}, hamming_loss : {hl}".format(mcc=mcc_RFC,acc = accuracy_RFC, hl = hamming_loss_RFC, i = i))
```

## Gradiente Boosting Classifier Tuning

```python
gbc = GradientBoostingClassifier(random_state=1)
```

```python
parameters = {
    "n_estimators":[5,50,250,500],
    "max_depth":[1,3,5,7,9],
    "learning_rate":[0.01,0.1,1,10,100]
}
```

```python
X_train, X_test, y_train, y_test = train_test_split(xi,yi, random_state=1670)

from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(gbc,parameters,cv=5)
cv.fit(X_train,y_train)
```

```python
def display(results):
    print(f'Best parameters are: {results.best_params_}')
    print("\n")
    mean_score = results.cv_results_['mean_test_score']
    std_score = results.cv_results_['std_test_score']
    params = results.cv_results_['params']
    for mean,std,params in zip(mean_score,std_score,params):
        print(f'{round(mean,3)} + or -{round(std,3)} for the {params}')
```

```python
display(cv)
```

```python
# tuning random_state
for z in tqdm(range(1000)):
    gbc = GradientBoostingClassifier(random_state=z,learning_rate=1,max_depth=9,n_estimators=250 )
    model_GBC = gbc.fit(X_train,y_train)
    pred_GBC = gbc.predict(X_test)
    accuracy_GBC = accuracy_score(y_test, pred_GBC)
    mcc_GBC = matthews_corrcoef(y_test, pred_GBC)
    hamming_loss_GBC = hamming_loss(y_test, pred_GBC)
    if accuracy_GBC > 0.65:
        print('accuracy: {ac}, mcc: {mcc}, hamming_loss: {hl}'.format(ac=accuracy_GBC,mcc=mcc_GBC,hl=hamming_loss_GBC))
```

```python
# tuning train test split
X_train, X_test, y_train, y_test = train_test_split(xi,yi, random_state = 10767)
gbc = GradientBoostingClassifier(random_state=1,learning_rate=1,max_depth=9,n_estimators=250 )
model_GBC = gbc.fit(X_train,y_train)
pred_GBC = gbc.predict(X_test)
accuracy_GBC = accuracy_score(y_test, pred_GBC)
mcc_GBC = matthews_corrcoef(y_test, pred_GBC)
hamming_loss_GBC = hamming_loss(y_test, pred_GBC)
print('accuracy: {ac}, mcc: {mcc}, hamming_loss: {hl}'.format(ac=accuracy_GBC,mcc=mcc_GBC,hl=hamming_loss_GBC))
```
