```python
import pandas as pd
import numpy as np
from pyADA import Similarity
import matplotlib.pyplot as plt
from tqdm import tqdm

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, matthews_corrcoef, roc_curve, auc, hamming_loss, classification_report, roc_auc_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
```

```python
pd.set_option('display.max_columns', None)
df_model = pd.read_csv('df_esteroids_MB_IN_AR.csv', sep=";")
#df_model.drop(['Unnamed: 0'], axis= 1, inplace=True)
df_model
```

```python
# separando grupo 1 e 2
df_grup1 =[]
df_grup2 =[]
for f in range(len(df_model.iloc[:,0])):
    if df_model.iloc[f,0]==2 or df_model.iloc[f,0]==0 or df_model.iloc[f,0]==4:
        df_grup1.append(df_model.iloc[f,:]) 
    if df_model.iloc[f,0]==0 or df_model.iloc[f,0]==5 or df_model.iloc[f,0]==1 or df_model.iloc[f,0]==3 or df_model.iloc[f,0]==6:
        df_grup2.append(df_model.iloc[f,:]) 
df_grup1 = pd.DataFrame(df_grup1)
df_grup2 = pd.DataFrame(df_grup2)
df_grup1
```

```python
# separando grupo 0, 1, 2 e 3 
yi = []
for f in range(len(df_model.iloc[:,0])):
    if df_model.iloc[f,0]==0:
        yi.append(0)
    if df_model.iloc[f,0]==2:
        yi.append(1)
    if df_model.iloc[f,0]==4:
        yi.append(2)
    if df_model.iloc[f,0]==5 or df_model.iloc[f,0]==1 or df_model.iloc[f,0]==3 or df_model.iloc[f,0]==6:
        yi.append(3) 
yi = pd.DataFrame(yi)
yi.value_counts()
```

```python
xi = df_grup1.iloc[:,1:27]
y = df_grup1.iloc[:,0]
xi
```

```python
y = pd.DataFrame(y)
y
```

```python
from sklearn.model_selection import train_test_split
xi_train, xi_test, y_train, y_test = train_test_split(xi,y,
                                                      test_size=0.25,
                                                      random_state=5)
print('Treino = {}, Test = {}'.format(len(xi_train), len(xi_test)))
```

```python
input_dimen = xi.shape[1] #lembrar disso, é a quantidade de colunas do xi

def create_network():
    net = Sequential()
    net.add(Dense(30, activation='relu',
               kernel_initializer='random_uniform', #método para inicializar os pesos
               input_dim = input_dimen)) #input_dim numero de xi
    net.add(Dense(units=128, activation='relu')) #camada oculta
    net.add(Dense(units=128, activation='relu'))
    net.add(Dense(units=1, activation='softmax')) #camada de saida "sigmoid" ou "softmax"
        
    opt = Adam(learning_rate=0.0001)
    #loss: "", "categorical_crossentropy"
    net.compile(loss = 'binary_crossentropy', optimizer = opt, metrics = ['accuracy'])
    
    return net
```

```python
# Tuning NN
from sklearn.model_selection import train_test_split
xi_train, xi_test, y_train, y_test = train_test_split(xi,y,
                                                      test_size=0.25,
                                                      random_state=1)
net = create_network()
history = net.fit(xi_train, y_train,
                 batch_size = 32,
                 epochs = 100)
```

```python
y_pred = net.predict(xi_test)
y_pred[y_pred <= 0.5] = 0
y_pred[y_pred > 0.5] = 1
y_pred
```
